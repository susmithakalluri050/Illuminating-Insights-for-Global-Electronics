```python
import pandas as pd
```

Reading all csv files

```python
stores=pd.read_csv("Stores.csv")
sales=pd.read_csv("Sales.csv")
products=pd.read_csv("Products.csv")
exchange_rates=pd.read_csv("Exchange_Rates.csv")
customers=pd.read_csv("Customers.csv",encoding='ISO-8859-1')
```

cleaning the data

```python
stores.head()
```

```python
stores.isnull().sum()
```

```python
stores.dtypes
```

```python
stores['Country']=stores['Country'].astype(str)
stores['State']=stores['State'].astype(str)
stores['Open Date']=pd.to_datetime(stores['Open Date'],format='%m/%d/%Y')
```

```python
stores.to_csv("cleaned_store_data.csv")
```

```python
sales.head()
```

```python
sales.isnull().sum()
```

```python
sales.dtypes
```

```python
sales['Order Date']=pd.to_datetime(sales['Order Date'],format='%m/%d/%Y')
sales['Delivery Date']=pd.to_datetime(sales['Delivery Date'],format='%m/%d/%Y')
sales['Currrency Code']=sales['Currency Code'].astype(str)
```

```python
sales.fillna('Not available',inplace=True)
sales
```

```python
sales.to_csv("cleaned_sales_data.csv")
```

```python
products.head()
```

```python
products.isnull().sum()
```

```python
products.dtypes
```

```python
products['Unit Cost USD'] = products['Unit Cost USD'].replace('[$,]', '', regex=True).astype(float)
products['Unit Price USD'] = products['Unit Price USD'].replace('[$,]', '', regex=True).astype(float)
```

```python
products.dtypes
```

```python
products.to_csv("cleaned_products_data.csv")
```

```python
customers.head()
```

```python
customers.isnull().sum()
```

```python
customers.dtypes
```

```python
customers['State Code'].fillna('NAN',inplace=True)
```

```python
customers.to_csv("cleaned_customer_data.csv")
```

```python
exchange_rates.head()
```

```python
exchange_rates.isnull().sum()
```

```python
exchange_rates.dtypes
```

```python
exchange_rates['Date']=pd.to_datetime(exchange_rates['Date'],format='%m/%d/%Y',errors='coerce')
```

```python
exchange_rates
```

```python
overall=pd.merge(sales,customers,on='CustomerKey',how='inner')
```

```python
overall.dropna(inplace=True)
```

```python
overall
```

```python
overall=pd.merge(overall,exchange_rates,left_on=['Order Date','Currency Code'],right_on=['Date','Currency'],how='inner')
```

```python
overall=pd.merge(overall,products,on='ProductKey',how='inner')
```

```python
overall=pd.merge(overall,stores,on=['Country','State','StoreKey'],how='inner')
```

```python
overall
```

```python
col=list(overall.columns)
col1=[]
for c in col:
    col1.append(c.replace(' ','_').lower())
    
overall.columns=col1
```

```python
col=list(stores.columns)
col1=[]
for c in col:
    col1.append(c.replace(' ','_').lower())
    
stores.columns=col1
```
